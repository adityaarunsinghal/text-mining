```python
# A function to detect date strings

def date_parser (string):
    new = str(string)
    new = new.strip()
    new = re.sub('[^a-zA-Z0-9]',' ', new)
    
    try:
        new = dateutil.parser.parse(new)
        return(new)
    except ValueError:
        return(string)
```

```python
newdf = pd.DataFrame()
for col in datadf:
    newdf[col] = datadf[col].apply(date_parser)
```

```python
lst = mist.iloc[:,1].dropna()[:100].sample(1800, replace=True)
lst = lst.append(mist.iloc[:,0].dropna()[:100].sample(200, replace=True)).reset_index(drop=True)
lst
```

```python
data = pd.read_csv('All Data/obsolete data/Lannet_test_names.csv')
```

```python
data = data[:2000]
data['word'] = lst
# data.to_csv('All Data/obsolete data/Lannet_test_names.csv')
```

```python
num_cols = data._get_numeric_data().columns
categorical = list(set(data.columns) - set(num_cols))
categorical
```

```python
%time
for col in categorical:
    counts = data[col].value_counts().to_dict()
    i = 0
    for each_category in counts.keys():
        i =i+1
        for every_other_category in list(counts.keys())[i:]:
            if not d.check(every_other_category):
                if ((stringdist.levenshtein(each_category, every_other_category)<=2) & (counts[each_category]/counts[every_other_category]>3)):
                    data[col] = data[col].str.replace(every_other_category,each_category)
```
